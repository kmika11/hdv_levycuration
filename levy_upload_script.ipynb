## 1. Add values from CSV file to JSON metadata template

```python
#load python modules
import json
import os
import csv
import pandas as pd
import requests
from pyDataverse.api import Api
```

```python
#import existing validated csv file
links = pd.read_csv('2012/valid_links.csv', header = None)
links.columns = ['ochre_link','filename']
links['filename'] = links['filename'].astype('string')
links['filename'].head()
```

```python
#add column with <a href =""> tags to display in dataset desciption as a link
links['html_link'] = 'OCHRE link: <a href = \"' + links['ochre_link'] + '\">' + links['ochre_link'] + '</a>'
links['html_link']
```

```python
links['filename'][0]
```

Thinking through the loop... 
#### For each row in links:

   1. open json template
   2. replace `['fields']['value': 'ID GOES HERE'` with `links[filename][row]`
   3. replace `['fields']['value': [{'dsDescriptionValue': {'value': 'OCHRE link: '` with `links['html_link'][row]`
   4. write to new json file

```python
#make sure for loop is reading correctly
for row in range(len(links)):
    print(links.loc[row, 'filename'])
```

```python
#write to directory 
for row in range(len(links)):
    with open('json/levy_template.json') as f:
        t = json.load(f)
    
        t['datasetVersion']['metadataBlocks']['citation']['fields'][0]['value'] = links.loc[row, 'filename']

        t['datasetVersion']['metadataBlocks']['citation']['fields'][3]['value'][0]['dsDescriptionValue']['value'] = links.loc[row, 'html_link']
        
        file = 'json/files/' + links.loc[row,'filename'] + '.json'
        with open(file, "w+") as out:
            json.dump(t, out)
```
